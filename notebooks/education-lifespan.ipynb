```python
import pandas as pd
import requests
import io
import chardet
import matplotlib.pyplot as plt
```

```python
education_url = "https://raw.githubusercontent.com/sjcasey21/linear-relationships/master/datasets/education-lifespan/Education.csv"
education_df = pd.read_csv(education_url, encoding='iso-8859-1', names=['country', 'education'])

lifespan_url = "https://raw.githubusercontent.com/sjcasey21/linear-relationships/master/datasets/education-lifespan/Lifespan.csv"
lifespan_df = pd.read_csv(lifespan_url, encoding='iso-8859-1', names=['country', 'lifespan'])
```

```python
df = pd.merge(education_df, lifespan_df, on='country', how='inner')
df['lifespan'] = pd.to_numeric(df['lifespan'].str.replace("[^\d.]", ""))
```

```python
df.plot(kind='scatter', x='education', y='lifespan')
```

```python
sample = df.sample(10)
sample.plot(kind='scatter', x='education', y='lifespan')
```

```python
sample2 = df.sample(10)
sample2.plot(kind='scatter', x='education', y='lifespan')
```
