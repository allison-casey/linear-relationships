```python
import pandas as pd
import requests
import io
import chardet
import matplotlib.pyplot as plt
```

```python
education_url = "https://raw.githubusercontent.com/sjcasey21/linear-relationships/master/datasets/education-lifespan/Education.csv"
education_df = pd.read_csv(education_url, encoding='iso-8859-1', names=['country', 'data'])

lifespan_url = "https://raw.githubusercontent.com/sjcasey21/linear-relationships/master/datasets/education-lifespan/Lifespan.csv"
lifespan_df = pd.read_csv(lifespan_url, encoding='iso-8859-1', names=['country', 'data'])
```

```python
df = pd.merge(education_df, lifespan_df, on='country', how='inner')
df['data_y'] = pd.to_numeric(df['data_y'].str.replace("[^\d.]", ""))
```

```python
df.plot(kind='scatter', x='data_x', y='data_y')
```
